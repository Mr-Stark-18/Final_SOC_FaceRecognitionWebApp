```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Graduate Admissions.csv')
```

```python
df.describe()
```

```python
df
```

```python
sns.lmplot(x='GRE Score',y='Chance of Admit ',data=df)
# There is a linear relation between GRE scores and chance of admit
```

```python
sns.lmplot(x='TOEFL Score',y='Chance of Admit ',data=df)
# There is a linear relation between TOEFL scores and chance of admit
```

```python
sns.boxplot(x='Research',y='GRE Score',data=df,palette='rainbow')
#we see that those with better research experience have better scores in GRE
```

```python
sns.boxplot(x='Research',y='TOEFL Score',data=df,palette='rainbow')
#we see that those with resaerch experience have better scores in TOEFL
```

```python
corr = df.corr()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
cmap = sns.diverging_palette(230, 20, as_cmap=True)
```

```python
sns.heatmap(corr, annot=True, mask = mask, cmap=cmap)
#marking the various correlations
```

```python
sns.violinplot(x="Research", y="GRE Score", data=df,palette='rainbow')
#students with better scores have reasearch experience
```

```python
sns.violinplot(x="Research", y="TOEFL Score", data=df,palette='rainbow')
#students with better scores generally have reasearch experience
```

```python
sns.violinplot(x="Research", y="Chance of Admit ", data=df,palette='rainbow')
#With research experience chance of Admit increases a lot
```

```python
sns.jointplot(x='CGPA',y='Chance of Admit ',data=df)
#It clearly shows that chance of admit increases with increase in CGPA
```
